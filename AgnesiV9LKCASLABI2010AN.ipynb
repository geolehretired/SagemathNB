### 1. Gegeben sind die beiden Funktionenscharen $g_a$und $w_a$ mit $g_a(x)=2 \cdot a \cdot e^{\frac{-x^2}{4\cdot a^2 }}$ und  $w_a (x)=\frac{8 \cdot a^3}{x^2+4 \cdot a^2 }$, $x∈R,a∈R^+$. 
<b>Die Graphen beider Funktionenscharen verlaufen sehr ähnlich.</b>,<br>
1.1$\;$Skizzieren Sie die Graphen von $g_a$ und $w_a$ für a = 2  im Intervall $[-4,4]$.


```python
import numpy as np
@interact
def interactive_2d_plotter1(a=(1,(0,10))):
    var('x')
    wa(x)=a**3*8/(x^2+4*a**2)
    ga(x)=2*a*np.exp(-x^2/(4*a**2))              
    W=plot(wa(x),x,-4,4,rgbcolor=(1,0,0),legend_label="$w_a$")
    G=plot(ga(x),x,-4,4,rgbcolor=(0,1,0),legend_label="$g_a$")              
    show(G+W,gridlines=True,legend_loc="best")
```

1.2	Zeigen Sie:<br> 
+ Die Graphen der Funktionsscharen von $g_a$ und $w_a$   haben den  gleichen relativen Hochpunkt   
+ die Graphen der Funktionsscharen von $g_a$ und $w_a$   sind achsensymmetrisch zur y-Achse
+ die Graphen der Funktionsscharen von $g_a$ und $w_a$   haben das gleiche asymptotische Verhalten

```python
%display latex
# haben den gleichen lokalen Hochpunkt:
var('x','a')
assume(a>0)
ga(x)=2*a*np.exp(-x^2/(4*a**2)) 
wa(x)=a**3*8/(x^2+4*a**2)
ga_1S(x)=diff(ga(x),x)
wa_1S(x)=diff(wa(x),x)
# Nullstellen beider Ableitungen
Nga=solve(ga_1S(x)==0,x)
Nwa=solve(wa_1S(x)==0,x)
wga=Nga[0].operands()[1]
wwa=Nwa[0].operands()[1]
display("Nullstelle\\;von\\;g_a':"+latex(wga))
display("Nullstelle\\;von\\;w_a':"+latex(wwa))
ga_2S(x)=diff(ga_1S(x),x)
wa_2S(x)=diff(wa_1S(x),x)
display("g_a''("+latex(wga)+")="+latex(ga_2S(wga))+"<0,\\;da\\; a>0")
display("w_a''("+latex(wwa)+")="+latex(wa_2S(wwa))+"<0,\\;da\\; a>0")
HPga=(wga,ga(wga))
HPwa=(wwa,wa(wwa))
display("also\\;sind\\;die\\;beiden\\;identischen\\;Nullstellen\\;\\;die x-Koordinaten\\;der\\;zugehörigen\\;lokalen\\;Hochpkte:P_1("+latex(wga)+"\\|"+latex(ga(wga))+")")
display("bzw. P_2("+latex(wwa)+"\\|"+latex(wa(wwa))+").")
```

```python
# die Graphen der Funktionsscharen von g_a und w_a  sind achsensymmetrisch zur y-Achse
display("g_a(-x)="+latex(ga(-x))+"=g_a(x)")
display("w_a(-x)="+latex(wa(-x))+"=w_a(x)")
```

```python
l1rechts=limit(ga(x),x=infinity)
l1links=limit(ga(x),x=-infinity)
display("\\lim_{x\\to\\infty}"+latex(ga(x))+"="+latex(l1rechts)+"=\\lim_{x\\to -\\infty}"+latex(ga(x)))
display("\\lim_{x\\to\\infty}"+latex(wa(x))+"="+latex(l1rechts)+"=\\lim_{x\\to -\\infty}"+latex(wa(x)))
```

```python
g2(x)=ga(x).subs(a=2)
display("g_2(x)="+latex(g2(x)))
w2(x)=wa(x).subs(a=2)
display("w_2(x)="+latex(w2(x)))
I1=integrate(g2(x),x,-4,4,hold=True)
I1w=np.round(float(integrate(g2(x),x,-4,4)),4)
display("I_1="+latex(I1)+"\\approx"+latex(I1w))
I2=integrate(w2(x),x,-4,4,hold=True)
I2w=np.round(float(integrate(w2(x),x,-4,4)),4)
display("I_2="+latex(I2)+"\\approx"+latex(I2w))
delta2=np.abs(I2w/I1w-1)*100
display("\\Delta_2=(\\frac{I_2}{I_1}-1)\\cdot 100="+latex(np.round(delta2,2))+"\\%")
delta1=np.abs(I1w/I2w-1)*100
display("\\Delta_1=(\\frac{I_1}{I_2}-1)\\cdot 100="+latex(np.round(delta1,2))+"\\%")
```

### 2.2	Um eine (noch) bessere Approximation des Integrals $\int_{-4}^{4}g_2(x)dx$   zu erhalten, verwendet man die Funktion  $w_{2,k}$ mit $w_{2,k}(x)=\frac{64}{k^2\cdot x^2+16}$.<br>
* Erläutern Sie die geometrische Bedeutung der Variablen k.<br>
* Ermitteln Sie denjenigen Wert von k für den $\int_{-4}^{4}g_2(x)dx$   und $\int_{-4}^{4}w_{2,k}(x)dx$  sich um weniger als $10^{-4}$  unterscheiden.


```python
var('x','k')
assume(k>=0)
w2k(x)=2**3*8/(k^2*x^2+4*2**2)
W2k=[]
for k in range(6):
    t=2**3*8/(k^2*x^2+4*2**2) 
    W2k.append(t)
p0=plot(W2k[0],x,-4,4)
p1=plot(W2k[1],x,-4,4)
p2=plot(W2k[2],x,-4,4)
p3=plot(W2k[3],x,-4,4)
p4=plot(W2k[4],x,-4,4)
p5=plot(W2k[5],x,-4,4)
show(p0+p1+p2+p3+p4+p5,gridlines=True,title="Darstellung von $w_{2,k}$")
   
```

<b><p><span style="color:green"> je gößer der Wert von k umso stärker "konzentriert sich der Graph" um die y-Achse, also erfolt eine Stauchung parallel zur x-Achse.</span> 

```python
var('x','k')
Ugl1text=abs(integrate(g2(x),x,-4,4,hold=True)-integrate(w2k(x),x,-4,4,hold=True))<=10^-4
display(latex(Ugl1text))
I1=float(integrate(g2(x),x,-4,4))
I2=integrate(w2k(x),x,-4,4)
ergk=find_root(abs(I1-I2)==0.00009,0,2)
display("k\\approx"+latex(ergk))
```

```python
#Veranschaulichung der Approximation
u=ergk
w2k(x)=2**3*8/(u**2*x^2+4*2**2)
g2(x)=2*2*np.exp(-x^2/(4*2**2))    
W=plot(w2k(x),x,-4,4,rgbcolor=(1,0,0),legend_label="$w_{2,k}$")
G=plot(g2(x),x,-4,4,rgbcolor=(0,1,0),legend_label="$g_2$")              
show(G+W,gridlines=True,legend_loc="best")
```

### 4.	Durch $y=\frac{1}{10}$ und den Graphen von $w_{0,5}$ wird der Querschnitt des Inneren einer Glocke definiert.  Durch $y=\frac{1}{10}$ und den Graphen von $w_{0,5}+0.05$  wird der Querschnitt des Äußeren der gleichen Glocke definiert. <br>
Ermitteln Sie das für die Glocke notwendige Materialvolumen und erläutern Sie Ihr Vorgehen.<br>

</span> 
### $V_{Material}=\pi \cdot \Big(\int_{0.1}^{g_a}(x_a(y)^2dy-\int_{0.1}^{g_i}(x_i(y))^2dy\Big)$

```python
w_innen(x)=8*(1/2)**3/(x^2+4*(1/2)**2)
g_i=w_innen(0)
display("g\\_i="+latex(g_i))
w_aussen(x)=w_innen(x)+5/100
g_a=w_aussen(0)
display("g\\_a="+latex(g_a))
```

```python
var('y')
gl_innen=y==w_innen(x)
display(latex(gl_innen))
erginnenterm=solve(gl_innen,x^2)
display(erginnenterm)
xquadinnen=erginnenterm[0].operands()[1]
display(latex(xquadinnen))
gl_aussen=y==w_aussen(x)
display(latex(gl_aussen))
ergaussenterm=solve(gl_aussen,x^2)
display(ergaussenterm)
xquadaussen=ergaussenterm[0].operands()[1]
display(latex(xquadaussen))
```

```python
VMaterialGlocke=np.pi*(integrate(xquadaussen,y,1/10,g_a,hold=True)-integrate(xquadinnen,y,1/10,g_i,hold=True))
VMaterialGlockeWert=np.pi*(integrate(xquadaussen,y,1/10,g_a)-integrate(xquadinnen,y,1/10,g_i))
display("V_{Material Glocke}="+latex(VMaterialGlocke)+"\\approx"+latex(np.round(float(VMaterialGlockeWert),2)))
```
